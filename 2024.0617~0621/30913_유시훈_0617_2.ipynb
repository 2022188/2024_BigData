```python
#라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
#한글 폰트 불러오기
plt.rc('font',family='Malgun Gothic')
#랜덤 라이브러리 불러오기
import random
#데이터 만들기
height = []
weight = []
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    
#제목 : 키와 몸무게의 상관관계
plt.title('키와 몸무게의 상관관계')

#선점도 그래프 그리기
plt.scatter(height,weight,c = height, cmap = 'RdPu')
#축 제목
plt.colorbar(label='키')
plt.xlabel('키')
plt.ylabel('몸무게')

plt.show()
```

```python
#라이브러리 불러오기 및 별명 지정하기
import matplotlib.pyplot as plt
#한글 폰트 불러오기
plt.rc('font',family='Malgun Gothic')
#랜덤 라이브러리 불러오기
import random
#데이터 만들기
height,weight,score = [],[],[]
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
    
#제목 : 키와 몸무게의 상관관계
plt.title('키와 몸무게의 상관관계')

#선점도 그래프 그리기
plt.scatter(height,weight,s = score,c = height,cmap = 'rainbow',alpha=0.5)
#축 제목
plt.colorbar(label='키')
plt.xlabel('키')
plt.ylabel('몸무게')

plt.show()
```

## 그래프 겹쳐 그리기
#### 꺾은선 그래프 그리기

```python
import matplotlib.pyplot as plt

singer = ['A','B','C','D','E']
week1 = [42,68,37,63,65]
week2 = [71,62,57,23,90]

#제목 오디션 프로그램 목표 현황
plt.title('오디션 프로그램 득표 현황')
#꺾은선형 겹쳐 그리기 범례 컬러
plt.plot(singer,week1)
plt.plot(singer,week2)
plt.legend()
plt.show()
```
